```python
import requests
from pprint import pprint
import pandas as pd
import os

```

```python
def get_data(cate_id):
        
    page_data = []

    no_results = False
    i = 0
    while not no_results:
        payload = {
            'source': 'amazon_search',
            'query': '*',  # Search for ""
            'start_page': 1+i*20,
            'pages': 20,
            'parse': True,
            'context': [
                {'key': 'category_id', 'value': cate_id}  # category id
            ],
        }
        print(payload)
        
        # Get response
        response = requests.request(
            'POST',
            'https://realtime.oxylabs.io/v1/queries',
            auth=('randir', 'uQbz7ufGtyiMeh_'),
            json=payload,
        )
        json_response = response.json()
        results = json_response['results']
        print(results)
        for result in results:
            if 'no_results' in result['content'] and result['content']['no_results']:
                no_results = True
                break
            products = result['content']['results']['organic']
            for product in products:
                
                page_data.append({
                    'title': product['title'],
                    'url_image': product['url_image'],
                    'url': product['url'],
                })
        i += 1
    return page_data
```

```python


for i in range(0, len(categories_ids)):
    # print(categories_titles[i])
    page_data = get_data(categories_ids[i])
    df = pd.DataFrame(page_data)
    df.to_csv('data\\' + categories_titles[i] + '.csv', index=False)
    # for i in range(0, df.shape[0]):
    #     img_data = requests.get(df.iloc[i]['url_image']).content
    #     with open('data\\covers\\' + str(i) +'.jpg', 'wb') as handler:
    #         handler.write(img_data)
```
